```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
inter_state = pd.read_csv("data/Inter-StateWarData_v4.0.csv",dtype={"BatDeath":"int32"})
intra_state = pd.read_csv("data/Intra-StateWarData_v4.1.csv",encoding="latin-1")
extra_state = pd.read_csv("data/Extra-StateWarData_v4.0.csv",encoding="latin-1",dtype={"BatDeath":"int32"})
non_state   = pd.read_csv("data/Non-StateWarData_v4.0.csv",encoding="latin-1")
intra_state = intra_state.astype({"SideADeaths":'int32',"SideBDeaths":"int32"})
intra_state["Deaths"] = intra_state["SideADeaths"] + intra_state["SideBDeaths"]
intra_state["Year"] = intra_state["StartYear1"]
inter_state["Deaths"] = inter_state["BatDeath"]
inter_state["Year"] = inter_state["StartYear1"]
extra_state["Deaths"] = extra_state["BatDeath"]
extra_state["Year"] = extra_state["StartYear1"]
non_state["Deaths"] = non_state["TotalCombatDeaths"]
non_state["Deaths"] = non_state["Deaths"].str.replace(",","").astype(int)
non_state["Year"] = non_state["StartYear"]
```

```python
fig, ax = plt.subplots(figsize=(15,7))
ax.hist(intra_state.groupby("WarNum").min()["StartYear1"], bins=40, label="Intra-State Wars",alpha=0.5)
ax.hist(extra_state.groupby("WarNum").min()["StartYear1"], bins=40, label="Extra-State Wars",alpha=0.5)
ax.hist(non_state.groupby("WarNum").min()["StartYear"], bins=40, label="Non-State Wars",alpha=0.5)
ax.hist(inter_state.groupby("WarNum").min()["StartYear1"], bins=40, label="Inter-State Wars",alpha=0.5)
ax.axvline(1939,c="k",alpha=0.7,dashes=[4,2])
ax.set_ylabel("Number of Wars")
ax.set_xlabel("Year")
ax.set_title("Frequency of Different Types of Wars in COW Dataset")
ax.legend()
plt.savefig("plots/freq.png")
```

```python
full = pd.DataFrame({"Year":range(1816,2008)}).set_index("Year")
full["Deaths"] = 0
for war_data in [inter_state,intra_state,extra_state,non_state]:
    war_data[war_data["Deaths"] < 0] = 0
    
    full["Deaths"] = full["Deaths"].add(war_data.groupby("Year").sum()["Deaths"],fill_value=0)
```

```python
fig, ax = plt.subplots(figsize=(15,7))
t = full.groupby(full.index//5).sum()
t.index = t.index * 5
ax.plot(t.index,t["Deaths"])
ax.axvline(1939,c="k",alpha=0.7,dashes=[4,2])
ax.fill_between(t.index,t["Deaths"],alpha=0.5)
ax.set_ylim(0,5000000)
ax.set_title("Total Number of Battle-related Casualties per Year in COW Dataset")
ax.set_xlabel("Year")
ax.set_ylabel("Battle-related Casualties")
plt.savefig("plots/deaths.png")
```

```python
war_data = pd.concat([inter_state,intra_state,extra_state,non_state],sort=True)

war_data = war_data.groupby("WarName").agg(pd.Series.mode)

war_data = war_data[["Year","WhereFought"]].drop(0)

def reduce(x):
    if type(x) == np.ndarray:
        return min(x)
    return x

war_data["Year"] = war_data["Year"].map(reduce)

pre = war_data[war_data["Year"] <= 1939]
post = war_data[war_data["Year"] > 1939]

pre = pre["WhereFought"].value_counts()
post = post["WhereFought"].value_counts()

pre = pre.sort_index().loc[:9]

post = post.sort_index().loc[:9]

pre = pre/pre.sum()
post = post/post.sum()
```

```python
colors = ["#ed6a5a","#f4f1bb", "#9bc1bc", "#5ca4a9", "#e6ebe0","#ECC8AF"]
fig, (ax1,ax2) = plt.subplots(1,2,figsize = (15,7))
ax1.pie(pre,labels=["W. Hemisphere", "Europe", "Africa", "Middle East" , "Asia", "Oceania"],autopct='%1.1f%%',colors=colors)
ax1.set_title("Before 9/1/1939")
ax2.pie(post,labels=["W. Hemisphere", "Europe", "Africa", "Middle East" , "Asia"],autopct='%1.1f%%',colors=colors[:-1])
ax2.set_title("After 9/1/1939")
fig.suptitle("Percentage of Wars Fought in Each Continent Before and After September 1st 1939 in COW Dataset")
plt.savefig("plots/pie.png")
```
